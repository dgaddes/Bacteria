```python
S16_set = []
record_ids = []
record_description = []
record_seq = []
for root, dirs, files in os.walk("./16S_Fasta", topdown=False):
    for name in files:
        print(os.path.join(root, name))
        S16_directory = './16S_Fasta/' + name
        record = SeqIO.read(S16_directory, "fasta")
        record_ids.append(record.id)
        record_description.append(record.description)
        record_seq.append(str(record.seq))
data = {'Description': record_description, 'Record id': record_ids, '16S Sequence': record_seq}
S16_set = pd.DataFrame(data=data)
S16_set.head(100)
```

```python
#Define Common Primers

#Forward Primers 
F8 = 'AGAGTTTGATCCTGGCTCAG'
F27 = 'AGAGTTTGATCMTGGCTCAG'
CC_F = 'CCAGACTCCTACGGGAGGCAGC'
F357F = 'CTCCTACGGGAGGCAGCAG'
F515F = 'GTGCCAGCMGCCGCGGTAA'
F533F = 'GTGCCAGCAGCCGCGGTAA'
F16S = 'CAACGAGCGCAACCCT'
F1237F = 'GGGCTACACACGYGCWAC'
V1F = 'GYGGCGNACGGGTGAGTAA'

#Reverse Primers
R519 = 'GWATTACCGCGGCKGCTG'
CDR = 'CTTGTGCGGGCCCCCGTCAATTC'
R907 = 'CCGTCAATTCMTTTRAGTTT'
R1100 = 'AGGGTTGCGCTCGTTG'
R1391 = 'GACGGGCGGTGTGTRCA'
R1492 = 'GGTTACCTTGTTACGACTT'
R1492 = 'ACCTTGTTACGACTT'
V6R = 'AGCTGACGACANCCATGCA'
```

```python
#Generate Amplicons of Bacterial Species from Primers

forward_primer = MutableSeq(F533F)
reverse_primer = MutableSeq(R1100)

fwd_tup = primer.degenerate_primer(forward_primer)
rev_tup = primer.degenerate_primer(reverse_primer)

calculated_data = primer.create_PCR_amplicon(S16_set, rev_tup, fwd_tup)
#calculated_data.to_csv('Bacterial_Amplicons.csv', sep='\t')
#calculated_data.head(5)
```

```python
#Generate Dendro of Bacterial Species from 16S Segments
file_name = "V4V7"
writen_alignment = PyClustalW.create_alignment_file(calculated_data, 'Amplicon', file_name)
writen_alignment
PyClustalW.ClustalW_alignment(str(file_name + '.faa'))
#ClustalW_alignment
PyClustalW.Phylo_tree(file_name)
```

```python
import pandas as pd
import Bio
from Bio import AlignIO
from Bio.Alphabet import generic_dna
from Bio.Seq import Seq
from Bio.SeqRecord import SeqRecord
from Bio.Align import MultipleSeqAlignment
from Bio.Align.Applications import ClustalwCommandline
from Bio import Phylo,SeqIO

alignment_list = []
min_legnth = calculated_data['Amplicon'].map(lambda x: len(x) if len(x) > 0 else 10**5).min()
file_name = 'V4V7_ABu'

for item in calculated_data['Record id']:
    if "Acinetobacter" in str(calculated_data.loc[(calculated_data['Record id'] == item)]['Description'].item()):
        print(calculated_data.loc[(calculated_data['Record id'] == item)]['Description'].item)
        if len(calculated_data.loc[(calculated_data['Record id'] == item)]['Amplicon'].item()) > 1:
            alignment_list.append(SeqRecord(Seq(calculated_data.loc[(calculated_data['Record id'] == item)]['Amplicon'].item()[1:min_legnth], generic_dna), 
            id = item,
            description = calculated_data.loc[(calculated_data['Record id'] == item)]['Description'].item()))
        else:
            pass
    else:
        pass  
writen_alignment = MultipleSeqAlignment(alignment_list)
AlignIO.write(writen_alignment, str(file_name +'.faa'), "fasta")

writen_alignment
PyClustalW.ClustalW_alignment(str(file_name + '.faa'))
#ClustalW_alignment
PyClustalW.Phylo_tree(file_name)
```
